```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import numpy as np

# Initialize the StandardScaler
scaler = StandardScaler()
```

```python
# Loop through each CSV file
for i in range(1, 13):
    # Load the CSV file into a DataFrame
    df = pd.read_csv(f'raw_data/{i}.csv', sep=";")
    
    #create cartesian columns from Lat, Lng with "convert_to_cartesian"
    #df['X'], df['Y'] = (6378137 *((df['Lng'] * np.pi) / 180) + 500000, 6378137 * np.log(np.tan(np.pi/4 + ((df['Lat'] * np.pi)/180)/2)))


    # Create delta_Lat and delta_Lng columns
    df['delta_Alt'] = df['Alt'].diff().fillna(0)

    #df['delta_X'] = df['X'].diff().fillna(0)
    #df['delta_Y'] = df['Y'].diff().fillna(0)

    df['delta_Lat'] = df['Lat'].diff().fillna(0)
    df['delta_Lng'] = df['Lng'].diff().fillna(0)


    # Drop the specified columns
    df = df.drop(columns=['Time', 'msDelta', 'Alt', 'Speed', 'AltGPS', 'HeadGPS'], errors='ignore')  # Drop specified columns

    # List of columns to normalize
    columns_to_minmax = ['Acc_X', 'Acc_Y', 'Acc_Z', 'Gyro_X', 'Gyro_Y', 'Gyro_Z', 'Tilt_X', 'Tilt_Y', 'Tilt_Z', 'Mag_X', 'Mag_Y', 'Mag_Z', 'Head', 'Temp', 'Press', 'delta_Alt']#, 'delta_Lat', 'delta_Lng']
    #columns_to_standarize = ['Alt', 'delta_Lat', 'delta_Lng', 'delta_Alt']

    # Defining the theoretical min-max values based on sensor specifications
    min_values = {
        'Acc_X': -16, 'Acc_Y': -16, 'Acc_Z': -16,
        'Gyro_X': -2000, 'Gyro_Y': -2000, 'Gyro_Z': -2000,
        'Tilt_X': -180, 'Tilt_Y': -180, 'Tilt_Z': -180,
        'Mag_X': -3000, 'Mag_Y': -3000, 'Mag_Z': -3000, 'Head': 0,
        'Temp': -40, 'Press': 300, 'delta_Alt': -5,# 'Alt': -1000,
        #'delta_Lat': -0.01, 'delta_Lng': -0.01
    }
    max_values = {
        'Acc_X': 16, 'Acc_Y': 16, 'Acc_Z': 16,
        'Gyro_X': 2000, 'Gyro_Y': 2000, 'Gyro_Z': 2000,
        'Tilt_X': 180, 'Tilt_Y': 180, 'Tilt_Z': 180,
        'Mag_X': 3000, 'Mag_Y': 3000, 'Mag_Z': 3000, 'Head': 360,
        'Temp': 85, 'Press': 1250, 'delta_Alt': 5,# 'Alt': 5000,
        #'delta_Lat': 0.01, 'delta_Lng': 0.01
    }

    for column in columns_to_minmax:
        df[column] = (df[column] - min_values[column]) / (max_values[column] - min_values[column])

    #df[columns_to_standarize] = scaler.fit_transform(df[columns_to_standarize])

    # Save the modified DataFrame back to the CSV file
    df.to_csv(f'data/clean_data/{i}_clean.csv', sep=";", index=False)
```
